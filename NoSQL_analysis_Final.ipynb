### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
#Converting the RatingValue to Integer
establishments.update_many(
    {},
    [
        {
            '$set': {
                'RatingValue': {
                    '$convert': {
                        'input': '$RatingValue',
                        'to': 'int',
                        'onError': None,  # Handle non-convertible values as None
                        'onNull': None    # Handle null values as None
                    }
                }
            }
        }
    ]
)
```

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query = {'LocalAuthorityName': {'$regex': 'London'}, 'RatingValue': {'$gte': 4}}
result_2 = establishments.find(query)

# Use count_documents to display the number of documents in the result
london_gte_4_documents = establishments.count_documents(query)

# Display the first document in the results using pprint
pprint(list(result_2.limit(1))) 
```

```python
# Convert the result to a Pandas DataFrame
results = list(establishments.find(query))
london_gte_4_documents_df = pd.DataFrame(results)

# Display the number of rows in the DataFrame
num_rows = len(london_gte_4_documents_df)
print(f'Number of rows in the DataFrame: {num_rows}')

# Display the first 10 rows of the DataFrame
london_gte_4_documents_df.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
#Converting the latitude and longitude to float
establishments.update_many(
    {},
    [
        {
            '$set': {
               'geocode.latitude': { '$toDouble': "$geocode.latitude" },
            'geocode.longitude': { '$toDouble': "$geocode.longitude" }
            }
        }
    ]
)
```

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

degree_search = 0.01
latitude = 51.49014200
longitude = 0.08384000

#Calculate the range
min_latitude = latitude - degree_search 
max_latitude = latitude + degree_search 
min_longitude = longitude - degree_search 
max_longitude = longitude + degree_search 

query = { "$or" : [ 
         {'RatingValue': {'$eq': 5}},
         {'geocode.latitude': {'$gte': min_latitude, '$lte': max_latitude}},
         {'geocode.longitude': {'$gte': min_longitude, '$lte': max_longitude}}
        ]
    }
    
sort = [('scores.Hygiene', 1)]
limit = 5

results_3 = list(establishments.find(query).sort(sort).limit(limit))

pprint(results_3)
```

```python
# Convert result to Pandas DataFrame
results_3_df = pd.DataFrame(results_3)
results_3_df
```
